```python
#pi chart
import matplotlib.pyplot as plt
# fig_1=plt.figure()
# ax=fig_1.add_axes([0,0,1,1])
fig_1,ax=plt.subplots()

langs=['english','french','spanish','latin','german']
students=[100,50,200,40,70]

_=ax.pie(students,labels=langs,autopct='%.1f%%')
ax.set_title('Language Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

v=np.linspace(0,10,20)
h=np.sin(v)
z=np.cos(v)
fig_3=plt.figure()
ax=fig_3.add_axes([0,0,1,1])
ax.scatter(v,h,color='red')
ax.scatter(v,z,color='blue')
plt.show()

```

#d scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt3=plt.figure()
ax=plt3.add_subplot(projection='3d')
o=20*np.random.random(100)#Generates 100 random numbers between 0 and 20
x=np.sin(o)
y=np.cos(o)
z = 20 * np.random.random(100)
ax.scatter(x,y,z,c=z,cmap='Greens')#cmap-->color map
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt3 = plt.figure()
ax = plt3.add_subplot(projection='3d')
o = 20 * np.random.random(100)
x = np.sin(o)
y = np.cos(o)
z = 20 * np.random.random(100)
ax.scatter(x, y, z, c=z, cmap='Greens')  # cmap-->color map

# Set the colors of the axes
ax.xaxis.label.set_color('red')  # Set x-axis label color to red
ax.yaxis.label.set_color('blue') # Set y-axis label color to blue
ax.zaxis.label.set_color('yellow') # Set z-axis label color to yellow

# Set the colors of the axis lines (ticks)
ax.tick_params(axis='x', colors='red')    # Set x-axis tick color to red
ax.tick_params(axis='y', colors='blue')   # Set y-axis tick color to blue
ax.tick_params(axis='z', colors='yellow')  # Set z-axis tick color to yellow


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt3=plt.figure()
ax=plt3.add_subplot(projection='3d')
o=20*np.random.random(100)#Generates 100 random numbers between 0 and 20
x=np.sin(o)
y=np.cos(o)
z = 20 * np.random.random(100)
ax.scatter(x,y,z,c=x,cmap='Greens')
ax.scatter(x,y,z,c=y,cmap='Reds') # Changed cmap to 'Reds'
ax.scatter(x,y,z,c=z,cmap='viridis')#cmap-->color map, changed 'yellows' to 'viridis'
plt.show()
```
