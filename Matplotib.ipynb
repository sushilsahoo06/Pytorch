Motplotlib
  . use for making plots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)

print(x)
```

```python
print(y)
```

```python
print(z)
```

ploting the data

```python
#sin curve
plt.plot(x,y)
plt.show()
```

```python
#cosin wave
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y,z)
plt.show()
```

```python
#adding title ,x-axis & y-axis
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin wave')
plt.show()
```

```python
#parabola
import numpy as np
import matplotlib.pyplot as plt

plt.ion()

x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-',x,np.cos(x),'r--')
plt.show()
```

bar plot


```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])# Adds axes to the figure
languages=['english','french','spanish','latin','german']
people=[100,50,200,40,70]
ax.bar(languages,people)
ax.set_xlabel('languages')
ax.set_ylabel('people')
plt.show()
```
