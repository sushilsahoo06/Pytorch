. Data visulazation library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#seaborn has some built-in datasets
tips=sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualisations the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
# This argument uses different marker sizes to represent different values in the 'size' column. The 'size' column likely represents the number of people in the dining party, so larger markers will indicate larger parties.
#This argument uses different marker styles (e.g., circles, squares) to represent different values in the 'smoker' column. This further helps in visually differentiating smokers and non-smokers.
#This argument uses different colors to represent different values in the 'smoker' column (e.g., 'Yes', 'No'). This allows you to visually distinguish between smokers and non-smokers in the plot.
#This argument creates separate subplots for different values in the 'time' column (e.g., 'Lunch', 'Dinner'). It essentially splits the data and creates a plot for each time category.
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.relplot(data=iris,x='sepal_length',y='sepal_width',hue='species',style='species',col='species',size='species')
#hue->>gives color
#style->>uses different marker styles (e.g., circles, squares)
```

Scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```
